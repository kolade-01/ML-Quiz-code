```python
df1 = df.drop(["date", "lights"], axis = 1)
```

```python
df1
```

To normalize the data to a scale

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(df1)
```

```python
normal = pd.DataFrame(scaler.fit_transform(df1), columns = df1.columns)
```

```python
normal
```

```python
features = normal.drop(columns = ["Appliances"])
```

```python
label = normal["Appliances"]
```

```python
x = features

y = label
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score, mean_absolute_error
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
y_test
```

```python
MAE = mean_absolute_error(y_test, prediction)
```

```python
round(MAE, 2)
```

```python
RSS = np.sum(np.square(y_test - prediction))
```

```python
round(RSS, 2)
```

```python
MSE = mean_squared_error(y_test, prediction)
```

```python
MSE
```

```python
RMSE = np.sqrt(MSE)
```

```python
round(RMSE, 3)
```

```python
R2 = r2_score(y_test, prediction)
```

```python
round(R2, 2)
```
